```python
#!/usr/bin/env python3
import re, sys, os
import pdfplumber

FIGURE_RE = re.compile(r'^\s*(figure|fig\.?|fig\s)', re.I)       # allow leading spaces
WORD_RE   = re.compile(r"[A-Za-zÀ-ÖØ-öø-ÿ']+")

def count_words(pdf_path: str) -> int:
    if not os.path.exists(pdf_path):
        raise FileNotFoundError(pdf_path)

    total = 0
    with pdfplumber.open(pdf_path, laparams={"all_texts": True}) as pdf:
        for page in pdf.pages:
            text = page.extract_text() or ''
            for line in text.splitlines():
                if FIGURE_RE.match(line):        # skip figure captions
                    continue
                total += len(WORD_RE.findall(line))
    return total

if __name__ == "__main__":
    pdf_file = (
        sys.argv[1]
        if len(sys.argv) == 2
        else ""
    )

    try:
        print(f"{count_words(pdf_file):,} words (figures excluded)")
    except Exception as exc:
        sys.exit(f"ERROR  {exc}")
```
